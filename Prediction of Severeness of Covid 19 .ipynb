```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
data= pd.read_csv('covid 19 dataset.csv')
def data_split(data, ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_size = int(len(data) * ratio)
    test_indices = shuffled[:test_size]
    train_indices = shuffled[test_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
np.random.permutation(8)
train, test = data_split(data, 0.2)
```

```python
X_train = train[['creatine kinase-MB, U/L', 'D-dimer, mg/L', 'Lactate dehydrogenase, U/L']].to_numpy()
X_test = test[['creatine kinase-MB, U/L', 'D-dimer, mg/L', 'Lactate dehydrogenase, U/L']].to_numpy()
```

```python
Y_train = train[['ICU/NN-ICU']].to_numpy().reshape(111 ,)
Y_test = test[['ICU/NN-ICU']].to_numpy().reshape(27 ,)
```

```python
clf = LogisticRegression()
clf.fit(X_train, Y_train)
```

```python
from flask import Flask, render_template, request

app = Flask(__name__)

@app.route("/")
def home():
    return render_template("home.html")

@app.route("/calculate", methods=["POST"])
def calculate():
    
    a = float(request.form["num1"])
    b = float(request.form["num2"])
    c = float(request.form["num3"])
    
    inputs = [a,b,c]
    infProb=clf.predict_proba([inputs])[0][1]
    infProb=round(infProb)
    
    if(infProb==1):
        condition='you condition is critical'
    else:
        condition="you are out of danger but adviced to stay at home"
    
   
    return render_template("result.html", total=condition)

if __name__ == "__main__":
    app.run()
```
